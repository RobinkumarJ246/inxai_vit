```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
dataset_path = "dtset6.csv"
smartdata = pd.read_csv(dataset_path)
```

```python
smartdata.head()
```

```python
smartdata.select_dtypes(include=["object","category",]).columns.to_list
```

```python
np.sum(smartdata.isnull().any(axis=1))
```

```python
plt.figure(figsize=(84,24))
sns.countplot(data=smartdata,x="category",palette="rocket")
```

```python
smartdata["category"].unique()
```

```python
smartdata.columns
```

```python
open_count = smartdata[smartdata['action'] == 'open'].shape[0]
close_count = smartdata[smartdata['action'] == 'close'].shape[0]
turn_on_count = smartdata[smartdata['action'] == 'turn_on'].shape[0]
turn_off_count = smartdata[smartdata['action'] == 'turn_off'].shape[0]

print(f"Count of 'open': {open_count}")
print(f"Count of 'close': {close_count}")
print(f"Count of 'turn_on': {turn_on_count}")
print(f"Count of 'turn_off': {turn_off_count}")
```

```python
plt.figure(figsize=(8, 8))
smartdata['category'].value_counts().plot.pie(autopct='%1.1f%%')
plt.ylabel('')
plt.title('Category Distribution')
```

```python
plt.figure(figsize=(8, 8))
smartdata['action'].value_counts().plot.pie(autopct='%1.1f%%')
plt.ylabel('')
plt.title('Action Distribution')
```

```python
plt.figure(figsize=(8, 8))
smartdata['action_needed'].value_counts().plot.pie(autopct='%1.1f%%')
plt.ylabel('')
plt.title('Action_Needed Distribution')
```
